```python
def getECOS(topic,freq,beg,end,item1='?',item2='?',item3='?',item4='?', key=ECOS_API_KEY, alias=None):
  url = 'https://ecos.bok.or.kr/api/StatisticSearch/'+key+'/json/kr/'+str(1)+'/'+str(1000)+'/'+topic+'/'+freq+'/'+beg+'/'+end \
          +'/'+item1+'/'+item2+'/'+item3+'/'+item4

  # SSL 버전 다운그레이드
  ssl_context = ssl.create_default_context()
  ssl_context.options &= ~ssl.OP_NO_SSLv3
  ssl_context.set_ciphers('DEFAULT:@SECLEVEL=1')

  # PoolManager 생성 시 ssl_context 전달
  http = urllib3.PoolManager(ssl_context=ssl_context)
  response = http.request('GET', url)

  assert response.status == 200

  contents = response.data.decode('utf-8')
  parsed_contents = json.loads(contents)['StatisticSearch']
  row_count = parsed_contents['list_total_count']

  if row_count > 0:
    if alias is None:
      first_row = parsed_contents['row'][0]
      columns_name = ''
      for i in range(4, 0, -1):
        if first_row[f'ITEM_NAME{i}'] is not None:
          columns_name = first_row[f'ITEM_NAME{i}']
          break
    else:
      columns_name = alias

    df = pd.DataFrame(parsed_contents['row'])[['TIME', 'DATA_VALUE']].set_index('TIME').rename(columns={'DATA_VALUE':columns_name})
    df.index = pd.PeriodIndex(df.index, freq=freq)
    df = df.astype(float)
  else:
    df = pd.DataFrame()
  return df
```
